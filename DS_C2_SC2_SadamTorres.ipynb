**a.** Extraer la información del archivo.

```python
import pandas as pd
import numpy as np

#extrayendo información del archivo con read_csv de pandas.
datos = pd.read_csv('/content/drive/MyDrive/Data_Science (Tec. Monterrey)/05_Pandas_Numpy/06_Reto/country_vaccinations.csv')

#imprimir primeras filas del dataframe.
print(datos.head())

```

**b.** Mostrar la estructura y tipos de datos de cada columna para identificar qué operaciones puedes realizar con cada una de ellas, asegurándote que las columnas con fechas sean del tipo datetime64.

```python
#ver la estructura y tipo de datos.
datos.info()
```

```python
#convertir columna 'date' a formato datetime64.
datos['date'] = pd.to_datetime(datos['date'])
```

**c.** Determinar la cantidad de vacunas aplicadas de cada compañía (con base en cómo lo reporta cada país en la columna vaccines, en otras palabras, agrupe por vaccines y realice la sumatoria).


```python
#agrupando por 'vaccines' con el método groupby() y realizando sumatoria sobre la columna 'daily_vaccinations' para obtener el total.
vacunas_por_compania = datos.groupby('vaccines')[['daily_vaccinations']].sum()

print(vacunas_por_compania)
```

**d.** Obtener la cantidad de vacunas aplicadas en todo el mundo.

```python
#sumar el total de la columna 'daily_vaccinations'
total_mundial = datos['daily_vaccinations'].sum()
print(f"Total de vacunas aplicadas en el mundo: {total_mundial}")
```

**e.** Calcular el promedio de vacunas aplicadas por país.

```python
#agrupar por país y calcular promedio sobre vacunas diarias.
promedio_pais = datos.groupby('country')[['daily_vaccinations']].mean()
print(promedio_pais)
```

**f.** Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo.

```python
#determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo
total_dia_esp = datos[datos['date'] == '2021-01-29']['daily_vaccinations'].sum()
print(f"Total de vacunas aplicadas el 29/01/21: {total_dia_esp}")
```

**g.** Crear un dataframe nuevo denominado conDiferencias que contenga los datos originales y una columna derivada (diferencias) con las diferencias de aplicación entre las columnas daily_vaccionations y daily_vaccionations_raw.

```python
#generar copia con copy()
conDiferencias = datos.copy()
#generando columna derivada 'diferencias'
conDiferencias['diferencias'] = conDiferencias['daily_vaccinations'] - conDiferencias['daily_vaccinations_raw']

print(conDiferencias[['date', 'diferencias']])
```

**h.** Obtener el periodo de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua.

```python
#utilizando los métodos max y min para obtener el registro más reciente y antiguo
periodo = datos['date'].max() - datos['date'].min()
print(f"El periodo es de: {periodo}")
```

**i.** Crear un dataframe nuevo denominado conCantidad que contenga los datos originales y una columna derivada (canVac) con la cantidad de vacunas utilizadas cada día (usar la columna vaccines y separar por el carácter , ).

```python
#crear dataframe 'conCantidad' columna derivada (canVac) separando por comas.
conCantidad = datos.copy()
#usar función lambda para dividir la cadena por ',' y contar los elementos
conCantidad['canVac'] = conCantidad['vaccines'].apply(lambda vacunas: len(str(vacunas).split(',')))

print(conCantidad[['vaccines', 'canVac']])
```

**j.** Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020.

```python
#generando dataframe 'antes20' con registros antes del 20 de diciembre de 2020
antes20 = datos[datos['date'] < '2020-12-20']
print(f'Registros encontrados: {len(antes20)}')
```

**k.** Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer.

```python
#buscar la cadena 'Pfizer' dentro de la columna 'vaccines'
pfizer = datos[datos['vaccines'].str.contains('Pfizer', case=False, na=False)]
print(f"Registros con Pfizer: {len(pfizer)}")
```

**l.** Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer) en un archivo de Excel denominado resultadosReto.xlsx, donde cada dataframe ocupe una hoja diferente.

```python
#almacenar los dataframes generados en un archivo Excel 'resultadosReto.xlsx'
with pd.ExcelWriter('resultadosReto.xlsx') as writer:

    conDiferencias.to_excel(writer, sheet_name='conDiferencias')
    conCantidad.to_excel(writer, sheet_name='conCantidad')
    antes20.to_excel(writer, sheet_name='antes20')
    pfizer.to_excel(writer, sheet_name='pfizer')
```
